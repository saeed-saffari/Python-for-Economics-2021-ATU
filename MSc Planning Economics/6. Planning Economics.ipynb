### Set sections

```python
int_demand = df1.iloc[0:7, 0:7]
int_demand
```

```python
fin_demand = df1.iloc[0:7, 8:-2]
fin_demand
```

### A matrix

```python
A = int_demand.div(total_output)
A = A.rename_axis('A')
A
```

### $I$ matrix

```python
I = np.eye(7)
I = pd.DataFrame(I)
I.index = A.index
I.columns = A.columns
I = I.rename_axis('I')
I
```

### $I - A$ matrix

```python
I_min_A = I - A
I_min_A = I_min_A.rename_axis('$I - A$')
I_min_A
```

### Leontief $(I-A)^{-1}$

```python
L = np.linalg.inv(I_min_A)
L = pd.DataFrame(L,index=I_min_A.index ,columns=I_min_A.columns)
L = L.rename_axis('$L$')
L
```

```python
entire = []
for i in L.columns:
    #print(L.loc[i,i])
    entire.append(L.loc[i,i])
    

print(entire)
```

```python
L['کشاورزی'].sum()
```

```python
TO = list(L.sum())
TO
```

```python
Outer = []
for i in L.columns:
    a = L[i].sum() - L.loc[i,i]
    Outer.append(a)

Outer
```

### <center> جدول متقارن اقتصاد ایران در سال 1390،فعالیت در فعالیت با تکنولوژی فعالیت 

```python
df3 = pd.read_excel('IO tables.xlsx', sheet_name=2, skiprows=1)
df3.head()
```

```python
df3.columns
```

```python
df3 = df3.set_index('نام رشته فعالیت ')
df3.head()
```

```python
int_demand = df3.iloc[0:6, 0:6]
int_demand
```

```python
fin_demand = df3.iloc[0:6, 7:-1]
fin_demand
```

```python
total_outputA = df3.iloc[-1, 0:6]
total_outputA
```

```python
total_outputB = df3.iloc[0:6,-1]
total_outputB
```

```python
FD = total_outputB - df3.iloc[0:6]['جمع تقاضای واسطه ']
FD
```

```python
V = df3.loc['ارزش افزوده ناخالص/محصول ناخالص داخلی(GDP)'][0:6]
V
```

## BL

```python
A = int_demand.div(total_outputA)

I = np.eye(6)
I = pd.DataFrame(I)
I.index = A.index
I.columns = A.columns

I_min_A = I - A

L = np.linalg.inv(I_min_A)
L = pd.DataFrame(L,index=I_min_A.index ,columns=I_min_A.columns)
L = L.rename_axis('$L$')
L
```

```python
A['کشاورزی ']
```

```python
A_prime = A.copy()
A_prime['کشاورزی '] = 0
A_prime
```

```python
I = np.eye(6)
I = pd.DataFrame(I)
I.index = A.index
I.columns = A.columns

I_min_A_p = I - A_prime

L_prime = np.linalg.inv(I_min_A_p)
L_prime = pd.DataFrame(L_prime,index=I_min_A_p.index ,columns=I_min_A_p.columns)
L_prime = L_prime.rename_axis('$L\_prime$')
L_prime
```

```python
L_mine = L - L_prime
L_mine
```

```python
BL = L_mine @ FD.values
BL
```

```python
BL_percent = BL.sum() / total_outputB.sum() *100
BL_percent
```

```python
print('BL')
for i in L_mine.columns:
    A_prime = A.copy()
    A_prime[i] = 0
    
    I = np.eye(6)
    I = pd.DataFrame(I)
    I.index = A.index
    I.columns = A.columns
    
    I_min_A_p = I - A_prime
    
    L_prime = np.linalg.inv(I_min_A_p)
    L_prime = pd.DataFrame(L_prime,index=I_min_A_p.index ,columns=I_min_A_p.columns)
    L_prime = L_prime.rename_axis('$L\_prime$')
    
    L_mine = L - L_prime
    
    BL = L_mine @ FD.values
    
    BL_percent = BL.sum() / total_outputB.sum() *100
    
    print('The share of {} sections is {:1.3f} percents.'.format(i, BL_percent))
```

## FL

```python
B = int_demand.divide(total_outputB, axis = 0)

I = np.eye(6)
I = pd.DataFrame(I)
I.index = B.index
I.columns = B.columns

I_min_B = I - B

M = np.linalg.inv(I_min_B)
M = pd.DataFrame(M,index=I_min_B.index ,columns=I_min_B.columns)
M = M.rename_axis('$M$')
M
```

```python
B_prime = B.copy()
B_prime.loc['کشاورزی'] = 0
B_prime
```

```python
#B = int_demand.div(total_outputB.values)

I = np.eye(6)
I = pd.DataFrame(I)
I.index = B.index
I.columns = B.columns

I_min_B = I - B_prime

M_prime = np.linalg.inv(I_min_B)
M_prime = pd.DataFrame(M_prime,index=I_min_B.index ,columns=I_min_B.columns)
M_prime = M_prime.rename_axis('$M\_prime$')
M_prime
```

```python
M_mine = M - M_prime
M_mine
```

```python
V
```

```python
FL = V.values @ M_mine
FL
```

```python
FL_percent = FL.sum() / total_outputB.sum() *100
FL_percent
```

```python
print('FL')
for i in M_mine.index:
    B_prime = B.copy()
    B_prime.loc[i] = 0
    
    I = np.eye(6)
    I = pd.DataFrame(I)
    I.index = B.index
    I.columns = B.columns
    
    I_min_B = I - B_prime
    
    M_prime = np.linalg.inv(I_min_B)
    M_prime = pd.DataFrame(M_prime,index=I_min_B.index ,columns=I_min_B.columns)
    M_prime = M_prime.rename_axis('$M\_prime$')
    
    M_mine = M - M_prime
    
    FL = V.values @ M_mine
    
    FL_percent = FL.sum() / total_outputB.sum() *100
    
    print('The share of {} sections is {:1.3f} percents.'.format(i, FL_percent))
```
