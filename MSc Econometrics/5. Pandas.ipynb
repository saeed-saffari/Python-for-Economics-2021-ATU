## Merge

```python
left = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A'  : ['A0', 'A1', 'A2', 'A3'],
    'B'  : ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k4'],
    'C'  : ['C0', 'C1', 'C2', 'C4'],
    'D'  : ['D0', 'D1', 'D2', 'D4']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how='inner', on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

# World Bank

```python
# pip install pandas-datareader
# conda install pandas-datareader
```

```python
from pandas_datareader import wb
```

```python
gdp = wb.download(indicator='NY.GDP.MKTP.CD',country='all', start=1960, end=2020)
```

```python
gdp.head()
```

```python
gdp.unstack().head()
```

```python
gdp2 = gdp.unstack()
```

```python
gdp2.columns = np.arange(1960, 2021)
```

```python
gdp2.head()
```

```python
gdp2.columns
```

```python
gdp3 = gdp2.T
gdp3.head()
```

```python
gdp3.index.rename('year', inplace=True)
```

```python
gdp3.head()
```

```python
#pd.set_option('display.max_rows',None)
#pd.reset_option('display.max_rows')
gdp3['Iran, Islamic Rep.'] = gdp3['Iran, Islamic Rep.'].interpolate()
gdp3['Iran, Islamic Rep.']
```

```python
gdp3['Iran, Islamic Rep.'].plot(figsize = (15,8), lw=3)

plt.show()
```

```python
gdp3['Iran, Islamic Rep.'].plot(figsize = (15,8), lw=3)
plt.title('Iran GDP', fontsize = 15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.grid()
plt.show()
```

```python
gdp3[['Iran, Islamic Rep.', 'Norway']].plot(figsize = (15,8), lw=3)
plt.title('Iran GDP', fontsize = 15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.grid()
plt.show()
```
